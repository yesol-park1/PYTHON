```python
# p.9 XML 파싱

import xml.etree.ElementTree as et

tree = et.ElementTree(file='menu.xml')
root = tree.getroot()

print(root.tag)
print(type(root))
```

```python
# p.10 XML 내용 프린팅

for child in root:
    print('tag:', child.tag, 'attributes:', child.attrib)
    for grandchild in child:
        print('\ttag:', grandchild.tag, 'attributes:', grandchild.attrib)
```

```python
# p.11 menu의 하위 태그 수와 breakfast의 item 수

print(len(root))
print(len(root[0]))

print(type(root[0]))
```

```python
# p.12 특정 태그의 속성 값 접근

print(root[0])
print(root[0].get("hours"))
print(root[0].keys())
print(root[0].items())
```

```python
# p.12 태그 이름으로 찾기 - find method

lunch = root.find("lunch")

print(lunch)
print(lunch.get("hours"))
print(lunch.keys())
print(lunch.items())
print(type(lunch))
```

```python
# p.13 같은 이름의 태그가 여러 개일 때 - findall method

print(root[0].findall("item"))
```

```python
# p.13 태그 안의 내용 출력 - .text

for item in root[0].findall("item"):
    print(item.get("price"))
    print(item.text)
```

```python
# p.17 json.dumps

import json

j1 = {"name": "홍길동", "birth": "0525", "age": 30}

print(j1)
print(json.dumps(j1))
```

```python
# p.18 json.dumps indent 옵션

print(json.dumps(j1, indent=2))

print(json.dumps([1, 2, 3]))
print(json.dumps((4, 5, 6)))
```

```python
# p.19 json.loads

j1 = {"name": "홍길동", "birth": "0525", "age": 30}

d1 = json.dumps(j1)

print(d1)
print(json.loads(d1))
```

```python
# p.20 myinfo.json 파일 만들기

myinfo = {
    "name": "홍길동",
    "birth": "0525",
    "age": 30
}

with open("myinfo.json", "w", encoding="utf-8") as f:
    json.dump(myinfo, f, ensure_ascii=False, indent=2)

print("myinfo.json 파일 생성 완료")
```

```python
# p.20 JSON 파일 로드

with open("./myinfo.json", encoding="utf-8") as f:
    data = json.load(f)

print(type(data))
print(data)
```

```python
# p.26 BeautifulSoup

from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen("http://pythonscraping.com/pages/page1.html")
bsObj = BeautifulSoup(html.read(), "html.parser")

print(bsObj.h1)
```

```python
# p.27 BeautifulSoup 태그 직접 접근

print(bsObj.h1)
print(bsObj.html.body.h1)
print(bsObj.body.h1)
print(bsObj.html.h1)
```

```python
# p.27 HTML 구조 예쁘게 출력

print(bsObj.html.prettify())
```
